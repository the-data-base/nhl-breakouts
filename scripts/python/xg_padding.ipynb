```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ratings_dumpp.csv")
df_f = df[df['primary_position_type'] == 'Forward']
df_d = df[df['primary_position_type'] == 'Defenseman']
```

```python
blah1 = df_d.groupby('player_name')['toi_m'].agg(avg_toi='mean').reset_index()
blah2 = df_f.groupby('player_name')['toi_m'].agg(avg_toi='mean').reset_index()

print(f"dmen: {blah1['avg_toi'].mean() / 3 / 60}")
print(f"fwd: {blah2['avg_toi'].mean() / 3 / 60}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = df_d.dropna(subset=['toi_m', 'adj_xga_ev_per60'])  # Drop rows with missing values in these columns

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['toi_m'], df['adj_xga_ev_per60'], marker='o', color='blue', label='Players')

# Add labels and title
plt.xlabel('TOI (minutes)')
plt.ylabel('xGA EV per 60')
plt.title('Adjusted xGA EV per 60 vs. TOI for Players')

# Add data labels for each point
for i, txt in enumerate(df['player_name']):
    plt.annotate(txt, (df['toi_m'].iloc[i], df['xga_ev_per60'].iloc[i]), textcoords='offset points', xytext=(0, 5), ha='center')

# Display the legend
plt.legend()

# Show the plot
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = df_f.dropna(subset=['toi_m', 'xga_ev_per60'])  # Drop rows with missing values in these columns

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['toi_m'], df['xga_ev_per60'], marker='o', color='blue', label='Players')

# Add labels and title
plt.xlabel('TOI (minutes)')
plt.ylabel('xGA EV per 60')
plt.title('xGA EV per 60 vs. TOI for Players')

# Add data labels for each point
for i, txt in enumerate(df['player_name']):
    plt.annotate(txt, (df['toi_m'].iloc[i], df['xga_ev_per60'].iloc[i]), textcoords='offset points', xytext=(0, 5), ha='center')

# Display the legend
plt.legend()

# Show the plot
plt.grid()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

metric = 'xga_ev_per60' 
toi = 'ev_toi_m'

df_viz = df
df_viz = df_viz.dropna(subset=[toi, metric])  # Drop rows with missing values in these columns

# Create an interactive scatter plot using Plotly
fig = px.scatter(df_viz, x=toi, y=metric, text='player_name', labels={toi: f'{toi}', metric: f'{metric}'},
                 title=f'{metric} vs. {toi} for Players')

# Customize the appearance
fig.update_traces(marker=dict(size=10, opacity=0.7))

# Add data labels for each point
fig.update_layout(annotations=[dict(x=df_viz[toi].iloc[i], 
                                    y=df_viz[metric].iloc[i], 
                                    text=df_viz['player_name'].iloc[i], 
                                    showarrow=False, 
                                    xanchor='center', 
                                    yanchor='bottom') for i in range(len(df_viz))])

# Show the interactive plot
fig.show()
```
